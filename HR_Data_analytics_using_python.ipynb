Identify departments with the highest attrition rate (Resigned %).

```python
df.head(2)
```

1. Group and Aggregate Data

```python
# Group by department and calculate the total and resigned employee counts
dept_counts = df.groupby('Department')['Status'].agg(
    total_emp='count',
    resigned=lambda x: (x == 'Resigned').sum()
)
dept_counts
```

```python
type(dept_counts)
```

2. Calculate the Attrition Rate Percentage

```python
# Calculate the attrition rate and sort the departments from highest to lowest
dept_counts['attrition_rate_%'] = (dept_counts['resigned'] / dept_counts['total_emp']) * 100
```

```python
# Display the final table, sorted by the attrition rate
print(dept_counts.sort_values("attrition_rate_%", ascending=False))
```
